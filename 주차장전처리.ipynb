# 데이터 전처리
### 주차장과 주정차 단속건수의 상관관계를 확인하기 위한 전처리 작업

```python
import pandas as pd
```

```python
df = pd.read_csv('서울시(동별)주차장_2020.csv')
df
```

```python
list = ['-']
df.isin(list).sum()
```

```python
df = df.replace('-', 0)
df.isin(list).sum()
```

```python
df = df[(df['자치구'] == '송파구') | (df['자치구'] == '영등포구')]
df
```

```python
df.dtypes
```

```python
df = df.astype({'합계':'int'})
df = df.astype({'합계.1':'int'})
df = df.astype({'건축물부설.2':'int'})
df = df.astype({'건축물부설.4':'int'})
df = df.astype({'건축물부설.3':'int'})
df = df.astype({'건축물부설.5':'int'})
```

```python
df
```

```python
df['합계'] = df['합계'] - df['건축물부설.2'] - df['건축물부설.4']
df['합계.1'] = df['합계.1'] - df['건축물부설.3'] - df['건축물부설.5']
df
```

```python
need_col = ['기간','자치구','동','합계','합계.1','합계.2','합계.3','합계.4','합계.5','노상','노상.1','노외','노외.1' ]

my_park = df[need_col]
my_park
```

```python
당산_index = my_park[my_park['동'].str.contains('당산')].index
양평_index = my_park[my_park['동'].str.contains('양평')].index
신길_index = my_park[my_park['동'].str.contains('신길')].index
대림_index = my_park[my_park['동'].str.contains('대림')].index
영등포_index = my_park[my_park['동'].str.contains('영등포')].index
풍납_index = my_park[my_park['동'].str.contains('풍납')].index
거여_index = my_park[my_park['동'].str.contains('거여')].index
마천_index = my_park[my_park['동'].str.contains('마천')].index
방이_index = my_park[my_park['동'].str.contains('방이')].index
송파_index = my_park[my_park['동'].str.contains('송파')].index
가락_index = my_park[my_park['동'].str.contains('가락')].index
문정_index = my_park[my_park['동'].str.contains('문정')].index
잠실_index = my_park[my_park['동'].str.contains('잠실')].index


my_park.loc[당산_index,'동'] = '당산동'
my_park.loc[양평_index,'동'] = '양평동'
my_park.loc[신길_index,'동'] = '신길동'
my_park.loc[대림_index,'동'] = '대림동'
my_park.loc[영등포_index,'동'] = '영등포동'
my_park.loc[풍납_index,'동'] = '풍납동'
my_park.loc[거여_index,'동'] = '거여동'
my_park.loc[마천_index,'동'] = '마천동'
my_park.loc[방이_index,'동'] = '방이동'
my_park.loc[송파_index,'동'] = '송파동'
my_park.loc[가락_index,'동'] = '가락동'
my_park.loc[문정_index,'동'] = '문정동'
my_park.loc[잠실_index,'동'] = '잠실동'

```

```python
int_col = [ '합계', '합계.1', '합계.2', '합계.3', '합계.4', '합계.5', '노상','노상.1', '노외', '노외.1']
```

```python
for col in int_col:
    my_park[col] = my_park[col].astype("str")
    my_park[col] = my_park[col].apply(lambda x: x.replace(",",''))
    my_park[col] = my_park[col].astype("int")
    
my_park
```

```python
str_col = ['기간','자치구']

my_park = my_park.groupby(['기간','자치구','동']).sum().reset_index()

my_park
```

```python
col_name = ['연도', '구', '동명', '주차장 개수', '주차장 면적', '민영 개수', '민영 면적', '공영 개수', '공영 면적', '노상 개수', '노상 면적', '노외 개수', '노외 면적']

my_park.columns = col_name
my_park = my_park.reset_index(drop=True)
my_park
```

```python
df = my_park.drop(index=[7,19], axis=0)
df
```
