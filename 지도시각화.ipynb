```python
df = pd.read_csv('전처리끝난데이터.csv', encoding= 'cp949')
```

```python
df = df.rename(columns={'동명':'EMD_KOR_NM'})
```

```python

# Choropleth 시각화
fig = px.choropleth(df, geojson=geo_json, locations='EMD_KOR_NM', color='주정차 단속건수',
                                color_continuous_scale='Blues',
                                featureidkey='properties.EMD_KOR_NM')
fig.update_geos(fitbounds="locations", visible=False)
fig.update_layout(title_text='개수', title_font_size=20)
```

```python
m = folium.Map([37.476281, 126.958369], tiles="OpenStreetMap", zoom_start=10)
folium.Choropleth(
    geo_data=geo_json, 
    data=df,
    columns=['구', '주정차 단속건수'],
    key_on='feature.properties.name', 
    fill_color='Blues', 
    fill_opacity=0.6,
    line_opacity=0.2,
    legend_name='단속건수 Rate (%)'
).add_to(m) 
folium.LayerControl().add_to(m) 
m
```

```python
for i in range(0, len(df)):
    latitude = df.iloc[i]['위도']
    longitude = df.iloc[i]['경도']
    location=(latitude, longitude)

    if df.iloc[i]['EMD_KOR_NM'] in ['낙성대동','위례동'] :
        color = 'white'
    else:
        color = '#3931cc'

    folium.CircleMarker(location, 
                        radius=10,
                        color=color,
                        fill_color=color,
                        fill_opacity = 0.1, 
                        opacity=0.0,  
                        popup=df.iloc[i]['EMD_KOR_NM'] + "\n" + str(int(round(df.iloc[i]['주정차 단속건수']/10000, 0)))
                       ).add_to(m)


folium.LayerControl(collapsed=False).add_to(m)

m
```
